# Facets

```python
!pip install facets-overview
```

```python
from facets_overview.generic_feature_statistics_generator import GenericFeatureStatisticsGenerator
from IPython.core.display import display, HTML
import base64
```

```python
# 生成統計信息的 proto
proto_combined = GenericFeatureStatisticsGenerator().ProtoFromDataFrames([{'name': 'Combined Data', 'table': sales}])
protostr_combined = base64.b64encode(proto_combined.SerializeToString()).decode("utf-8")

# 創建 HTML 内容
html_content_combined = f'''
    <script src="https://cdnjs.cloudflare.com/ajax/libs/webcomponentsjs/1.3.3/webcomponents-lite.js"></script>
    <link rel="import" href="https://raw.githubusercontent.com/PAIR-code/facets/master/facets-dist/facets-jupyter.html">
    <facets-overview id="elem_combined"></facets-overview>
    <script>document.querySelector("#elem_combined").protoInput = "{protostr_combined}"</script>
'''
# 顯示 Facets Dive
display(HTML(html_content_combined))
```

```python
train = pd.read_csv('/content/titanic/train.csv')
test = pd.read_csv('/content/titanic/test.csv')

msno.bar(train)
```

```python
# Visualizing the locations of the missing data
msno.matrix(train)
```

```python
msno.matrix(train.sample(100))
```

```python
#sorted by Age
sorted = train.sort_values('Age')
msno.matrix(sorted)
```

```python
# Finding reason for missing data using a Heatmap
msno.heatmap(train)
```

```python
# Finding reason for missing data using Dendrogram
msno.dendrogram(train)
```
